Resumeix gràficament el data set DelayedFlights.csv

Crea almenys una visualització per:

Una variable categòrica (UniqueCarrier)\
Una variable numèrica (ArrDelay)\
Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)\
Dues variables numèriques (ArrDelay i DepDelay)\

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('DelayedFlights.csv')
```

```python
# Visualització per una variable categòrica (UniqueCarrier)
```

```python
x = df["UniqueCarrier"].value_counts()

plt.hist(x)
plt.show()
# NO ÉS EN ABSOLUT EL QUE NECESSITO, vull que apareixin les aerolínies. PASSEM A COLUMNES
```

```python
x = ['9E', 'AA', 'AQ', 'AS', 'B6', 'CO', 'DL', 'EV', 'F9', 'FL', 'HA', 'MQ', 'NW', 'OH',
  'OO', 'UA', 'US', 'WN', 'XE', 'YV']
y = [51885, 191865, 750, 39293, 55315, 100195, 114238, 81877, 28269, 71284, 7490,
  141920, 79108, 52657, 132433, 141426, 98425, 377602, 103663, 67063]
plt.title('2018 Flights by Airline', fontsize = 12, fontweight = 'bold')
plt.barh(x, y, color = "hotpink")
plt.show()
```

```python
plot = df['UniqueCarrier'].value_counts().plot(kind='barh', title='Flights/Airline', color='r')
plt.show()  
```

```python
# Visualització per una variable numèrica (ArrDelay)
```

```python
y = df['ArrDelay']
#trobar millor representació de 1 variable numèrica
#l´eix x s´autocrea amb els index dels valors de y
plt.ylabel('ArrDelay')
plt.grid(axis='y', color='yellow')

plt.plot(y)
plt.show()
```

```python
# Visualització d'una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
```

```python
y = df['UniqueCarrier']
x = df['ArrDelay']

plt.title('Arribal Delays by Airline')
plt.xlabel('ArrDelay')
plt.ylabel('UniqueCarrier')
plt.scatter(x, y, color = 'g', alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(14, 18))
sns.boxplot(x=df['UniqueCarrier'], y=df['ArrDelay'])
plt.show()
```

```python
# Visualització de dues variables numèriques (ArrDelay i DepDelay)
```

```python
x = df['ArrDelay']
y = df['DepDelay']

plt.xlabel('ArrDelay')
plt.ylabel('DepDelay')
plt.title('Delay Correlation Arrivals vs Departures')
plt.scatter(x, y, c = 'blue', alpha=0.1)
plt.show()
```

```python
# Visualització de tres variables (ArrDelay, DepDelay i UniqueCarrier)
```

```python
data = df[['UniqueCarrier', 'DepDelay', 'ArrDelay']]
title = 'Delays Relation by Airline'
sns.relplot(data=data, x='DepDelay', y='ArrDelay', hue='UniqueCarrier')
plt.title(title, fontsize=20, color='#6146C6', alpha=0.8, pad=50)
plt.show()
```
